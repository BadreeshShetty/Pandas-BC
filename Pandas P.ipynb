__Group__ cars by the columns __model_year and origin__ and return the __mean mpg__ for each group <br>
__Column labels__ of mpg_year_origin shall be __europe__, __japan__ & __usa__ 

```python
mpg_year_origin = cars.groupby(["model_year", "origin"]).mpg.mean().unstack().round(2)
```

```python
mpg_year_origin
```

```python
mpg_year_origin.plot()
plt.show()
```

It seems that manufacturer from europe, japan and usa were able to improve efficiency

__Group__ cars by __model_year and origin__ and return __mean__, __max__ and __min mpg__ for all groups

```python
cars.groupby(["model_year", "origin"]).mpg.agg(["mean", "min", "max"]).unstack().round(2)
```

return the columns __name__ and __mpg__ for the __two most fuel efficient cars__ for __each combination of model_year & origin__

__create__ the __user defined function__ get_most_efficient

```python
def get_most_efficient(group):
    return group.nlargest(n =2, columns = "mpg").loc[:, ["name", "mpg"]]
```

__Apply__ get_most_efficient on the appropriate __groupby object__

```python
most_eff = cars.groupby(["model_year", "origin"]).apply(get_most_efficient)
```

```python
most_eff = most_eff.droplevel(-1)
```

```python
most_eff.head(10)
```

__Select__ the 2 most efficient cars from __japan__ in __1980__

```python
most_eff.loc[(80, "japan")]
```

The most efficient car is... the mazda glc.

Calculate the __mean mpg__ for each combination of __model_year & origin__ and __assign__ the corresponding __group-specific value__ to all cars. Then, __filter__ all cars where the __absolute difference__ between __mpg__ and __group-specific mpg__ is __greater than 10__.

First, __group__ cars by __model_year & origin__ and calculate the __mean mpg__ for each group via the --- method to create the new column __"group_mpg"__

```python
cars["group_mpg"] = cars.groupby(["model_year", "origin"]).mpg.transform("mean").round(2)
```

__Inspect!__ The group-specific mpg for the vw pickup is ... ?

```python
cars.head()
```

```python
cars.tail()
```

The group-specific mpg for the vw pickup is... 40.0!

__Create__ the column __"mpg_outlier"__ by __substracting__ the __group_mpg__ column from the __mpg__ column.

```python
cars["mpg_outlier"] = (cars.mpg-cars.group_mpg).round(2)
```

```python
cars.mpg_outlier.describe()
```

__Filter__ cars for all cars/rows, where the __absolute value__ in the __mpg_outlier__ column is __greater than 10__

Cars with the lowest __fuel efficiency__ come from?

```python
plt.figure(figsize=(12,8))
sns.set_palette("pastel")
sns.violinplot(data = cars, x = "origin", y = "mpg", dodge = True)
sns.swarmplot(data = cars, x = "origin", y = "mpg", dodge = True, color = "black")
plt.show()
```

Cars with the lowest __fuel efficiency__ come from... usa.

Do you think that the lower fuel efficiency of us cars is __statistically significant__? 

```python
plt.figure(figsize=(12,8))
sns.set_palette("pastel")
sns.barplot(data = cars, x = "origin", y = "mpg", dodge = True)
plt.show();
```

Yes, it seems to be statistically significant.

Do you think that manufacturers from __all three regions significantly increased fuel efficiency__ from 1970 till 1982?

```python
plt.figure(figsize=(12,8))
sns.set_palette("pastel")
sns.pointplot(data = cars, x = "model_year", y = "mpg", hue = "origin", dodge = True, ci = 95)
plt.show();
```

Yes, they all increased fuel efficiency significantly.

```python
plt.figure(figsize=(12,8))
sns.set_palette("pastel")
sns.pointplot(data = cars, x = "model_year", y = "displacement", hue = "origin", dodge = True, ci = 95)
plt.show();
```

Manufacturers from the usa managed to increase fuel efficiency by reducing engine power (horsepower, displacement) and weight. Manufacturers from europe and japan were able to increase fuel efficiency without reducing engine power and weight.
